```python
import torch 
import torch.nn.functional as F

import numpy as np 
import matplotlib.pyplot as plt
from sklearn.utils import shuffle

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using {} device".format(device))
```

```python
# Parameters of training
learning_rate = 0.001
epsilon = 1e-8

num_iter = 10000
batch_size = 256

test_ratio = 0.1
num_networks = 5

num_print = 1000
test_size = 10
```

# Making Toy Data
Toy dataset is y = 10sin(x) + epsilon

If x < 0:  = N(0, sigma1^2) -> In this example sigma1 = 3 <br>
If x > 0:  = N(0, sigma2^2) -> In this example sigma2 = 1

```python
data_range = 7
data_step = 0.001

bound1 = -2
bound2 = 2

data_sigma1 = 0.1
data_sigma2 = 0.5
num_data = 1

data_x1 = np.arange(-data_range, bound1 + data_step, data_step)
data_x2 = np.arange(bound2, data_range + data_step, data_step)
data_x = np.concatenate((data_x1, data_x2))
data_x = np.reshape(data_x, [data_x.shape[0], 1])

data_y = np.zeros([data_x.shape[0], 1])

data_y1_true = np.sin(data_x1)
data_y2_true = np.sin(data_x2)

for i in range(data_x.shape[0]):
    if (data_x[i,0] < bound1): 
        data_y[i, 0] = np.sin(data_x[i,0]) + np.random.normal(0, data_sigma1)
    else:
        data_y[i, 0] = np.sin(data_x[i,0]) + np.random.normal(0, data_sigma2)

data_x, data_y = shuffle(data_x, data_y)
        
num_train_data = int(data_x.shape[0] * (1 - test_ratio))
num_test_data  = data_x.shape[0] - num_train_data

train_x = data_x[:num_train_data, :]
train_y = data_y[:num_train_data, :]
test_x  = data_x[num_train_data:, :]
test_y  = data_y[num_train_data:, :]

print("Train data shape: " + str(train_x.shape))
print("Test data shape: " + str(test_x.shape))

plt.plot(train_x, train_y, 'b*', markersize=1)
plt.plot(test_x, test_y, 'y*', markersize=1)
plt.plot(data_x1, data_y1_true, 'r')
plt.plot(data_x2, data_y2_true, 'r')
plt.legend(['Data', 'y=x^3'], loc = 'best')
plt.title('y = sin(x) + $\epsilon$')
plt.show()
```

```python
# Make batch data 
def make_batch(data_size, sample_size, data_x, data_y):
    # Make batches(testing)
    batch_idx = np.random.choice(data_size, sample_size)
    
    batch_x = np.zeros([sample_size, num_data])
    batch_y = np.zeros([sample_size, 1])
        
    for i in range(batch_idx.shape[0]):
        batch_x[i,:] = data_x[batch_idx[i], :]
        batch_y[i,:] = data_y[batch_idx[i], :] 
    
    batch_x = torch.FloatTensor(batch_x).to(device)
    batch_y = torch.FloatTensor(batch_y).to(device)
    
    return batch_x, batch_y   

def to_np(x):
    return x.cpu().detach().numpy()
```

# Model

```python
class model(torch.nn.Module):
    def __init__(self):
        super(model, self).__init__()
        self.fc1 = torch.nn.Linear(1, 16)
        self.fc2 = torch.nn.Linear(16, 32)
        self.fc_mu = torch.nn.Linear(32, 1)
        self.fc_sig = torch.nn.Linear(32, 1)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        mu = self.fc_mu(x)
        sig = self.fc_sig(x)
        
        return mu, sig
```

```python
model_list = []
opt_list = []

for i in range(num_networks):
    model_list.append(model().to(device))
    opt_list.append(torch.optim.Adam(model_list[i].parameters(), lr = learning_rate))
```

```python
train_data_num = train_x.shape[0]
test_data_num  = test_x.shape[0]

loss_train = np.zeros([num_networks])
out_mu     = np.zeros([test_size, num_networks])
out_sig    = np.zeros([test_size, num_networks])

for iter in range(num_iter):
    # Make batches(testing)
    batch_x_test, batch_y_test = make_batch(test_data_num, test_size, test_x, test_y)
        
    for i in range(num_networks):
        # Make batches(training)
        batch_x, batch_y = make_batch(train_data_num, batch_size, train_x, train_y)
       
        # Training
        mu_train, sig_train = model_list[i](batch_x)
        sig_train_pos = torch.log(1 + torch.exp(sig_train)) + 1e-6
        
        loss = torch.mean(0.5*torch.log(sig_train_pos) + 0.5*(torch.square(batch_y - mu_train)/sig_train_pos)) + 1
    
        if np.any(np.isnan(to_np(loss))):
            print(0.5*torch.log(sig_train_pos))
            print(0.5*(torch.square(batch_y - mu_train)/sig_train_pos))
            raise ValueError('There is Nan in loss')
            
        opt_list[i].zero_grad()
        loss.backward()
        opt_list[i].step()
        
        # Testing
        mu_test, sig_test = model_list[i](batch_x_test)
        sig_test_pos = torch.log(1 + torch.exp(sig_test)) + 1e-6
        
        loss_train[i] += loss.item()
        out_mu[:, i] = np.reshape(to_np(mu_test), (test_size))
        out_sig[:, i] = np.reshape(to_np(sig_test_pos), (test_size))
        
    # Get final test result
    out_mu_final = np.mean(out_mu, axis = 1)
    out_sig_final = np.sqrt(np.mean(out_sig, axis=1) + np.mean(np.square(out_mu), axis = 1) - np.square(out_mu_final))
    
    if iter % num_print == 0 and iter != 0:
        print(('-------------------------') + ' Iteration: ' + str(iter) + ' -------------------------')
        print('Average Loss(NLL): ' + str(loss_train / num_print))
        print('x: ' + str(batch_x[0]))
        print('mu: ' + str(out_mu[0, :]))
        print('std: ' + str(np.sqrt(out_sig[0, :])))
        print('Final mu: ' + str(out_mu_final[0]))
        print('Final std: ' + str(out_sig_final[0]))
        print('Real Value: ' + str(batch_y_test[0]))
        print('\n')
        
        loss_train = np.zeros(num_networks)
```

# Testing with Known and Unknown Data 


```python
# Get Unknown dataset and test 
x_sample = np.arange(-12, 12 + 0.1, 0.1)
x_sample = np.reshape(x_sample, [x_sample.shape[0], 1])

y_sample = np.zeros([x_sample.shape[0], 1])

for i in range(x_sample.shape[0]):
    y_sample[i, 0] = np.sin(x_sample[i,0])

x_sample_tensor = torch.FloatTensor(x_sample).to(device)
y_sample_tensor = torch.FloatTensor(y_sample).to(device)
    
# output for ensemble network
out_mu_sample  = np.zeros([x_sample_tensor.shape[0], num_networks])
out_sig_sample = np.zeros([x_sample_tensor.shape[0], num_networks])

# output for single network
out_mu_single  = np.zeros([x_sample_tensor.shape[0], 1])
out_sig_single = np.zeros([x_sample_tensor.shape[0], 1])

for i in range(num_networks):
    mu_sample, sig_sample = model_list[i](x_sample_tensor)
    sig_sample = torch.log(1 + torch.exp(sig_sample)) + 1e-6
    
    mu_sample = to_np(mu_sample)
    sig_sample = to_np(sig_sample)
    
    out_mu_sample[:,i]  = np.reshape(mu_sample, (x_sample_tensor.shape[0]))
    out_sig_sample[:,i] = np.reshape(sig_sample, (x_sample_tensor.shape[0]))

    out_mu_single[:,0]  = np.reshape(mu_sample[:,0], (x_sample_tensor.shape[0]))
    out_sig_single[:,0] = np.reshape(sig_sample[:,0], (x_sample_tensor.shape[0]))

out_mu_sample_final  = np.mean(out_mu_sample, axis = 1)
out_sig_sample_final = np.sqrt(np.mean(out_sig_sample, axis=1) + np.mean(np.square(out_mu_sample), axis = 1) 
                               - np.square(out_mu_sample_final))
out_sig_sample_aleatoric = np.sqrt(np.mean(out_sig_sample, axis=1))
out_sig_sample_epistemic = np.sqrt(np.mean(np.square(out_mu_sample), axis = 1) - np.square(out_mu_sample_final))

plt.figure(1)
plt.plot(data_x, data_y, 'b*', markersize=1, label='Data')
plt.plot(x_sample, y_sample, 'r', linewidth = 2, label='y=sin(x)')
plt.plot(x_sample, out_mu_sample_final - out_sig_sample_final, 'c', linewidth = 2)
plt.plot(x_sample, out_mu_sample_final + out_sig_sample_final, 'c', linewidth = 2)

plt.legend()
plt.title('y = sin(x) + $\epsilon$ where $\epsilon$ ~ N(0, {}^2) and N(0, {}^2)'.format(data_sigma1, data_sigma2))

plt.show()
```
